```python
from pjs_qlab.data.YahooPriceFetcher import YahooPriceFetcher as price_fetcher
```

```python
from pypfopt import HRPOpt
from pypfopt import expected_returns
```

```python
import pandas as pd
```

```python
tickers=['NUE','COST','UNH','ORCL','LLY','DE','CAT','BRK-B','AVGO','ABBV']
```

```python
yObj=price_fetcher(tickers,period='2y')
```

```python
closes=yObj.get_close()
```

```python
returns=expected_returns.returns_from_prices(closes)
```

```python
returns
```

```python
hrp=HRPOpt(returns)
hrp.optimize()
```

```python
weights=hrp.clean_weights()
weights
```

```python
portfolio_values=closes.dot(pd.Series(weights))
portfolio_values
```

```python
weights_df=pd.DataFrame(list(weights.items()),columns=['ticker','weight'])
weights_df
```
